```python
import pandas as pd
import string


import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer


nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('punkt_tab')


stopwords.words('english')
exclude = string.punctuation
```

```python
df = pd.read_csv('/content/reviews_dataset.csv')
df.shape
```

```python
df.head()
```

```python
df['news'] = df['news'].str.lower()
def clean_and_tokenize(text):
  text = re.sub(r'[^a-zA-Z ]+', '', text)
  tokens = word_tokenize(text)
  return tokens

df['news'] = df['news'].apply(clean_and_tokenize)
```

```python
df.head()
```

```python
def lemmatize_tokens(tokens):
  lemmatizer = WordNetLemmatizer()
  return [lemmatizer.lemmatize(token) for token in tokens]

df['news'] = df['news'].apply(lemmatize_tokens)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
X = df['news']
Y = df['type']

print(X)
print(Y)

encoder = LabelEncoder()
Y = encoder.fit_transform(Y)

print(Y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
X_train = [str(x) for x in X_train]
X_test = [str(x) for x in X_test]
vectorizer = CountVectorizer()
X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators=150, random_state=42)
rf_classifier.fit(X_train_vectorized, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix
y_pred = rf_classifier.predict(X_test_vectorized)
accuracy = accuracy_score(y_test, y_pred)
print("Overall Accuracy:", accuracy)
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
precision = precision_score(y_test, y_pred, average=None)
print("Class-wise Precision:", precision)
```
